```python
#7.>
df4=pd.DataFrame({'Email_adress':['asutosh@domain.com','bunty@domain.com','sudh@domain.com','krish@domain.com','alakh@domain.com','amitabh@domain.com','jisu@domain.com','bola@domain.com','kporad@domain.com','nulu@domain.com','mosad@domain.com']})
def username(x):
    a=list(x)
    b=''
    for i in a:
        if i=='@':
            break
        b+=i
    return(b)
df4['username']=df4['Email_adress'].apply(username)
print(df4)
        
```
