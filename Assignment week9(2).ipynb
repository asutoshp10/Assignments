Matplotlib is a plotting library for the Python programming language. It is used to create graph and helps in data visualizations in Python. The various plots that can be created using Pyplot are:

Line Plot,
Histogram,
Scatter,
3D Plot,
Pie chart,Bar graph

```python
#2.> The plot in which the data points are not connected to each other are called scatter plot
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter graph')
plt.show()
```

```python
#3.>Subplot function is used to plot multiple graph  at a time
import numpy as np
import matplotlib.pyplot as plt
#plot 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

#plot 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

#plot 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

#plot 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
"""4.> a bar plot is a representation of a bar graph or a chart that represents data using rectangular 
bars. These plots are commonly used to compare the values of different categories or groups. They are 
useful for displaying data that is grouped into discrete categories"""
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
#5.> 
#box 1
box1 = np.random.normal(100, 10, 200)
plt.subplot(1,2,1)
plt.boxplot(box1)

#box 2
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()
```
